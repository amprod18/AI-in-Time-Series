# Custom ODE

```python
import torch

def custom_ODE(t, x):
    # x = x.unsqueeze(0)
    true_A = torch.tensor([[-0.1, 2.0], [-2.0, -0.1]])
    return torch.mm((x.unsqueeze(0))**3, true_A).squeeze(0)

spiral_parameters = {'ode_func':custom_ODE, 'start_t':0., 'max_t':25., 'n_points':2000, 'look_ahead':1, 'initial_guess':Tensor([2., 0.]), 'method':'rk4', 'ode_params':{}}
spiral_dataset = create_synthetic_dataset(spiral_parameters, display_heads=True, plot_result=True)
```
